```python
import datetime
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from statsmodels.tsa.seasonal import seasonal_decompose
from pandas.plotting import register_matplotlib_converters
from pyhive import hive
```

```python
#database connection
conn = hive.Connection(host="192.168.1.241", port=10000, username="hive")
```

```python
#use command
cursor = conn.cursor()
cursor.execute("use RoadNetwork")
```

```python
#defining constant variables
fTag1 = "t1.flowvalue"
fTag2 = "t2.flowvalue"
dTag1 = "t1.measurementdatetime"
dTag2 = "t2.measurementdatetime"
sTag1 = "t1.speedvalue"
sTag2 = "t2.speedvalue"
kTag1 = "a.density"
kTag2 = "b.density"
timeRange = "2"
```

```python
#testing variables
dateTimeFrom = "2020-02-03 00:00:00.0"
dateTimeTo = "2020-02-04 00:00:00.0"
referenceA = "RWS01_MONIBAS_0021hrl0580ra"
referenceB = "RWS01_MONIBAS_0021hrr0582ra"
```

```python
#reading data set from database
stmt = "SELECT t1.measurementdatetime, t1.flowvalue, t1.speedValue, t2.measurementdatetime, t2.flowvalue, t2.speedValue FROM measurementpoint t1, measurementpoint t2  WHERE t1.measurementdatetime >= t2.measurementdatetime - interval '" + timeRange + "' minute AND t1.measurementdatetime <= t2.measurementdatetime + interval '" + timeRange + "' minute AND t1.reference = '" + referenceA + "' AND t2.reference = '" + referenceB + "' AND t1.measurementdatetime < '" + dateTimeTo + "' AND t1.measurementdatetime > '" + dateTimeFrom + "' AND t2.measurementdatetime < '" + dateTimeTo + "' AND  t2.measurementdatetime  > '" + dateTimeFrom + "' ORDER BY t1.measurementdatetime"
data = pd.read_sql(stmt, conn)
```

```python
data.head()
```

```python
#getting descriptive statistical values for the flow data
flowMean1 = data.mean()[0] #mean
flowSDeviation1 = data.std()[0] #standard deviation
#dividing dataset into quantiles
flowInterval1_1 = data.quantile(0.25)[0]
flowInterval1_2 = data.quantile(0.50)[0]
flowInterval1_3 = data.quantile(0.75)[0]
flowInterval1_4 = data.quantile(1)[0]
```

```python
#getting descriptive statistical values for the flow data
flowMean2 = data.mean()[2] #mean
flowSDeviation2 = data.std()[2] #standard deviation
#dividing dataset into quantiles
flowInterval2_1 = data.quantile(0.25)[2]
flowInterval2_2 = data.quantile(0.50)[2]
flowInterval2_3 = data.quantile(0.75)[2]
flowInterval2_4 = data.quantile(1)[2]
```

```python
#defining mean variables for each of the quantile intervals
flowMean1_1 = 0
flowMean1_2 = 0
flowMean1_3 = 0

#as the sets may be unequally sized we define element count variables as well
n1_1 = 0
n1_2 = 0
n1_3 = 0

#getting the means of the three different intervals
    
    #note that to do the means the zero values aren't considered 
    
for index, row in data.iterrows():
    
    if (row[fTag1] > flowInterval1_1 and row[fTag1] <= flowInterval1_2):
        flowMean1_1 = flowMean1_1 + row[fTag1]
        n1_1 = n1_1 + 1
    elif (row[fTag1] > flowInterval1_2 and row[fTag1] <= flowInterval1_3):
        flowMean1_2 = flowMean1_2 + row[fTag1]
        n1_2 = n1_2 + 1
    elif (row[fTag1] > flowInterval1_3 and row[fTag1] <= flowInterval1_4):
        flowMean1_3 = flowMean1_3 + row[fTag1]
        n1_3 = n1_3 + 1

flowMean1_1 = flowMean1_1 / n1_1
flowMean1_2 = flowMean1_2 / n1_2
flowMean1_3 = flowMean1_3 / n1_3
```

```python
#defining mean variables for each of the quantile intervals
flowMean2_1 = 0
flowMean2_2 = 0
flowMean2_3 = 0

#as the sets may be unequally sized we define element count variables as well
n2_1 = 0
n2_2 = 0
n2_3 = 0

#getting the means of the three different intervals
    
    #note that to do the means the zero values aren't considered 
    
for index, row in data.iterrows():
    
    if (row[fTag2] > flowInterval2_1 and row[fTag2] <= flowInterval2_2):
        flowMean2_1 = flowMean2_1 + row[fTag2]
        n2_1 = n2_1 + 1
    elif (row[fTag2] > flowInterval2_2 and row[fTag2] <= flowInterval2_3):
        flowMean2_2 = flowMean2_2 + row[fTag2]
        n2_2 = n2_2 + 1
    elif (row[fTag2] > flowInterval2_3 and row[fTag2] <= flowInterval2_4):
        flowMean2_3 = flowMean2_3 + row[fTag2]
        n2_3 = n2_3 + 1

flowMean2_1 = flowMean2_1 / n2_1
flowMean2_2 = flowMean2_2 / n2_2
flowMean2_3 = flowMean2_3 / n2_3
```

```python
#cleaning data
for index, row in data.iterrows():
    #first check if the zero value must be changed
        #the zero values which have the previous neighbor value above the first quantile will be changed
    if (row[fTag1] == 0):
        if (index > 0): #preventing from index out of bounds error
            if (data.iloc[[index-1][0]][fTag1] > flowInterval1_2): #checking if the value is extremly low
                if (data.iloc[[index-1][0]][fTag1] < flowInterval1_3): #checking the quantile which the previous neighbor belongs
                    data.set_value(index, fTag1, flowMean1_2)
                else:
                    data.set_value(index, fTag1, flowMean1_3)
    if (row[fTag2] == 0):
        if (index > 0): #preventing from index out of bounds error
            if (data.iloc[[index-1][0]][fTag2] > flowInterval2_2): #checking if the value is extremly low
                if (data.iloc[[index-1][0]][fTag2] < flowInterval2_3): #checking the quantile which the previous neighbor belongs
                    data.set_value(index, fTag2, flowMean2_2)
                else:
                    data.set_value(index, fTag2, flowMean2_3)
```

```python
#plotting data from the two measurement points
plt.plot(data[dTag1], data[fTag1])
plt.plot(data[dTag1], data[fTag2])
plt.xlabel("Date Time")
plt.ylabel("Traffic Flow")
plt.rcParams['figure.figsize'] = [25,10]
plt.show()
```

```python
#print correlation between flows 
data[fTag1].corr(data[fTag2])
```

```python
#calculate densities
densityDF = pd.DataFrame(columns = [kTag1, kTag2])

for index, row in data.iterrows():
    dA = row[fTag1] / row[sTag1]
    dB = row[fTag2] / row[sTag2]
    auxDF = pd.DataFrame([[dA, dB]], columns = [kTag1, kTag2])
    densityDF = densityDF.append(auxDF, ignore_index = True)
```
