```python
#constant variables
fTag = "flowvalue"
sTag = "speedvalue"
dTag = "measurementdatetime"
kTag = "density"
```

```python
def backward_function(root_id, q_dict, kc_dict, q_origin, kc_origin):
    q = q_origin
    for key in kc_dict:
        kc_total = 0
        if (key != root_id):
            for i in kc_dict[key]:
                kc_total = kc_total + i
            kc_total = kc_total + kc_origin
            q = q - (kc_origin / kc_total)*q_dict[key]
    kc_total = 0
    for i in kc_dict[root_id]:
        kc_total = kc_total + i
    kc_total = kc_total + kc_origin
    q = q * (kc_total/kc_origin)
    return q
```

```python
def forward_function(q_dict, kc_dict, kc):
    q = 0
    for key in q_dict:
        kc_total = 0
        for i in kc_dict[key]:
            kc_total = kc_total + i
        kc_total = kc_total + kc
        q = q + (kc / kc_total) * q_dict[key]
    return q
```

```python
def timeToInt(root, strTime):
    #taking the train flow dataframe as the default one to transform a date into an integer
    time = t.strptime(strTime, "%H:%M:%S")
    time = datetime.fromtimestamp(mktime(time)).time()
    mb = root.getMB()
    trainFlowDF = mb.getTrainFlowDF()
    timeInt = trainFlowDF.shape[0] - 1
    for index, row in trainFlowDF.iterrows():
        auxTime = row[dTag]
        if (auxTime >= time):
            timeInt = index
            return timeInt
    return timeInt
```

```python
def first_backward(root):
    mb = root.getMB()
    kc = mb.getCriticalDensity() # critical density without the congestion
    kc_new = kc * p # critical density with the congestion
    flow_model = mb.getFlowModel()  
    q = flow_model(timeToInt(root, strTime)) # flow without congestion
    q_new = q * p # flow with congestion
    root.setNewCriticalDensity(kc_new)
    root.setNewFlow(q_new) 
    for i in root.getAdjacentRoots():
        if (i.getMaster() < 0):
            backward(i, root)
        if (i.getMaster() > 0):
            forward(i)
```

```python
def first_forward(root, timeInt):
    mb = root.getMB()
    flow_model = mb.getFlowModel()  
    q = flow_model(timeInt) # flow without congestion
    q_new = q * p # flow with congestion
    root.setNewFlow(q_new)
    for i in root.getAdjacentRoots():
        if (i.getMaster() < 0):
            backward(i, root)
        if (i.getMaster() > 0):
            forward(i)
```

```python
def backward(root, root_origin):
    
    q_dict = {}
    kc_dict = {}
    
    if (isinstance(root.getMB(), str) == False):
        
        for i in root_origin.getAdjacentRoots():
            if (i.getMaster() < 0 and i != root): 
                
                if (i.getNewFlow() > 0):
                    q_dict[i.getNewFlow()] = i.getNewFlow()
                else:
                    mb_i = i.getMB()
                    flow_model_i = mb_i.getFlowModel()
                    q_i = flow_model_i(timeToInt(i, strTime)) 
                    q_dict[i.getNodeId()] = q_i
                    
                kc_i_list = []
                for j in i.getAdjacentRoots():
                    if (j.getMaster() > 0 and j != i):
                        mb_j = j.getMB()
                        kc_j = mb_j.getCriticalDensity()
                        kc_i_list.append(kc_j)
                kc_dict[i.getNodeId()] = kc_i_list
                
        kc_i_list = []
        for i in root.getAdjacentRoots():
            if (i != root_origin and i.getMaster() > 0):
                mb_i = i.getMB()
                kc_i = mb_i.getCriticalDensity()
                kc_i_list.append(kc_i)
        kc_dict[root.getNodeId()] = kc_i_list
        
        if (root_origin.getNewCriticalDensity() >= 0): # the flow of the origin node must a new one 
            q_new = backward_function(root.getNodeId(), q_dict, kc_dict, root_origin.getNewFlow(), root_origin.getNewCriticalDensity())
        else:
            q_new = backward_function(root.getNodeId(), q_dict, kc_dict, root_origin.getNewFlow(), root_origin.getMB().getCriticalDensity())
        root.setNewFlow(q_new)
        
        for i in root.getAdjacentRoots():
            if (i.getMaster() < 0 and i != root_origin):
                backward(i, root)
            if (i.getMaster() > 0 and i != root_origin):
                forward(i)
                
    else:
        
        for i in root.getAdjacentRoots():
            if (i.getMaster() < 0 and i != root_origin):
                backward(i, root_origin)
            if (i.getMaster() > 0 and i != root_origin):
                forward(i)
```

```python
def forward(root):
    
    mb =  root.getMB()
    
    if (isinstance(mb, str) == False):
        
        kc = mb.getCriticalDensity()
        kc_dict = {}
        q_dict = {}
        
        for i in root.getAdjacentRoots():
            
            if (i.getMaster() < 0):
                kc_i_list = []
                for j in i.getAdjacentRoots():
                    if (j.getMaster() > 0 and i != j):
                        kc_i_list.append(j.getCriticalDensity())
                kc_dict[i.getNodeId()] = kc_i_list
                if (i.getNewFlow() > 0):
                    q_dict[i.getNodeId()] = i.getNewFlow()
                else:
                    mb_i = i.getMB()
                    flow_model_i = mb_i.getFlowModel()
                    q_dict[i.getNodeId()] = flow_model_i(timeToInt(i, strTime)) #new flow
        
        root_origin = root.getOrigin()
        kc_i_list = []
        for i in root_origin.getAdjacentRoots():
            if (i.getMaster() > 0):
                mb_i = i.getMB()
                kc_i_list.append(mb_i.getCriticalDensity())
        print("\n------\n" + kc_i_list + "\n------\n")
        kc_dict[i.getNodeId()] = kc_i_list
        if (root_origin.getNewFlow() > 0):
            q_dict[root_origin.getNodeId()] = root_origin.getNewFlow()
        else: 
            flow_model_origin = root_origin.getFlowModel()
            q_origin = flow_model_origin(timeToInt(root_origin, strTime))
            q_dict[root_origin.getNodeId()] = q_origin
        
        print(str(root.getNodeId()) + ", " + str(q_dict))            
        q_new = forward_function(q_dict, kc_dict, kc)
        root.setNewFlow(q_new)
        
        for i in root.getAdjacentRoots():
            if (i.getMaster() < 0):
                backward(i, root)
            if (i.getMaster() > 0):
                forward(i)
    else:
        
        for i in root.getAdjacentRoots():
            if (i.getMaster() < 0):
                backward(i, root)
            if (i.getMaster() > 0):
                forward(i)
```
